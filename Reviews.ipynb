```python
# Creating a function for Review Rating
def change1(value):
  if value=='10/10':
    return 'positive'
  else:
    return value
```

```python
def change2(value):
  if value >= '5/10':
    return 'positive'
  else:
    return 'negative'
```

```python
revi['Review Rating']=revi['Review Rating'].apply(change1)
```

```python
revi['Review Rating']=revi['Review Rating'].apply(change2)
```

```python
# Start Preprocessing for Review body
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# variable for wordnetlem
```

```python
wl=WordNetLemmatizer()
```

```python
corpus=[]

for i in range(0,len(review)):
    rev=re.sub('[^a-zA-Z]',' ',revi['Review title'][i])
    rev=rev.lower()
    rev=rev.split()

    rev=[wl.lemmatize(word) for word in rev if not word in stopwords.words('english')]
    rev=' '.join(rev)
    corpus.append(rev)
    
```

```python
import tensorflow_hub as hub
```

```python
import tensorflow_text as text
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Embedding
```

```python
# Download Basic Bert Model from hub
```

```python
bert_preprocess = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3")
bert_encoder = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-24_H-1024_A-16/4")
```

```python
# def getpro(senti):
#   pro_text=bert_preprocess(senti)
#   return bert_encoder(pro_text)['pooled_output']
```

#Model Creation

```python

# Bert layers
text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
preprocessed_text = bert_preprocess(text_input)
outputs = bert_encoder(preprocessed_text)
# Neural network layers
l = tf.keras.layers.Dropout(0.1, name="dropout")(outputs['pooled_output'])
l = tf.keras.layers.Dense(1, activation='sigmoid', name="output")(l)
# Use inputs and outputs to construct a final model
model = tf.keras.Model(inputs=[text_input], outputs = [l])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
x_new=corpus.copy()
```

```python
y_new=Y.values.reshape(-1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_new,y_new,test_size=0.3,random_state=45)
```

```python
x_n=np.array(x_train)
x_t=np.array(x_test)
```

```python
model.fit(x_n,y_train,epochs=3,batch_size=32)
```

```python
y_pred=model.predict(x_t)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(classification_report(y_test,y_pred.round()))
```
